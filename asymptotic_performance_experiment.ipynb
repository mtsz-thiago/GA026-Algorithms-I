# Asymptotic Performance Experiment

This notebook allows the empiric verification of algorithms performance by executing multiple instances of the problem and fiting the experimental data using regressions

```python
from measure.base import *
from algorithms.sort import insert_sort

import numpy as np
import matplotlib.pyplot as plt
```

```python
elapsed = [r.elapsed for r in output]
```

```python
plt.figure()
plt.hist(elapsed)
plt.title("Elapsed time histogram")
plt.xlabel("elpased time [s]")
plt.ylabel("#")
plt.grid(True)
```
